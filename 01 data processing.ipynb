### 2.3 裁剪动作

```python
import os
lis = os.listdir("cut_video/")
cutlis = [each.split(".")[0] for each in lis]
```

```python
for i in trange(len(namelis)):
    clear_output()
    name = namelis[i]
    
    cut = train[train["name"] == name][train["class"] >= 41][train["class"] < 49]
    for eachind in cut.index:
        if name+"_"+str(cut.loc[eachind,"begin"]) in cutlis:
            continue
        clip = VideoFileClip("video/"+name+".mp4").subclip(cut.loc[eachind,"begin"],cut.loc[eachind,"end"])
        clip.write_videofile("cut_video/"+name+"_"+str(cut.loc[eachind,"begin"])+".mp4",fps = 20)
        cutlis.append(name+"_"+str(cut.loc[eachind,"begin"]))
```

```python
from tqdm import trange
```

```python
subcutlis = []
classlis = []
```

```python
lis = os.listdir("cut/")
subcutlis = [each.split(".")[0] for each in lis]
```

```python
for i in trange(len(cutlis)):
    clear_output()
    name = cutlis[i]
    
    faname = name.split("_")[0]
    cut = train[train["name"] == faname][train["class"] >= 41][train["class"] < 49][train["begin"] == int(name.split("_")[1])]
    for eachind in cut.index:
        if name+"_"+str(cut.loc[eachind,"subbegin"]) in subcutlis:
            continue
        if cut.loc[eachind,"class"] < 46:
            clip = VideoFileClip("cut_video/"+name+".mp4").subclip(cut.loc[eachind,"subbegin"],cut.loc[eachind,"subend"]+1)
            clip.write_videofile("val_cut/"+name+"_"+str(cut.loc[eachind,"subbegin"])+".mp4",fps = clip.fps)
        else:
            clip = VideoFileClip("cut_video/"+name+".mp4").subclip(cut.loc[eachind,"subbegin"],cut.loc[eachind,"subend"]+1.5)
            clip.write_videofile("val_cut/"+name+"_"+str(cut.loc[eachind,"subbegin"])+".mp4",fps = clip.fps)
        subcutlis.append(name+"_"+str(cut.loc[eachind,"subbegin"]))
        classlis.append(cut.loc[eachind,"class"])
```

### 2.4 写出label

```python
df = pd.DataFrame(columns = ["class"])
```

```python
for name in subcutlis:
    faname = name.split("_")[0]
    cut = train[train["name"] == faname][train["begin"] == int(name.split("_")[1])][train["subbegin"] == int(name.split("_")[2])]
    df.loc[name] = cut["class"].values.tolist()[0]
```

```python
df.to_csv("label.txt")
```

### 2.5 整理成需要的格式

```python
import numpy as np
```

```python
cut = pd.read_csv("label.txt",index_col = 0)
```

```python
cut.index = [each+".mp4" for each in cut.index]
```

```python
train_ind = np.random.choice(cut.index,int(np.floor(len(cut.index)*0.8)))
```

```python
test_ind = [each for each in cut.index if each not in train_ind]
```

```python
from shutil import copyfile
import os
```

```python
try:
    os.mkdir("train/")
    os.mkdir("val/")
except:
    pass
```

```python
for each in train_ind:
    try:
        copyfile("cut/"+each, "train/"+each)
    except:
        print("Unable to copy file. %s" % each)
```

```python
for each in test_ind:
    try:
        copyfile("cut/"+each, "val/"+each)
    except:
        print("Unable to copy file. %s" % each)
```

```python
train = cut.loc[train_ind]
test = cut.loc[test_ind]
```

```python
train.index = ["../mydata/train/"+each for each in train.index]
test.index = ["../mydata/val/"+each for each in test.index]
```

```python
train.to_csv("mydata_train.txt",sep = ' ',header = None)
test.to_csv("mydata_val.txt",sep = ' ',header = None)
```
